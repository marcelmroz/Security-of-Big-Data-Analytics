Correlation means the relation between variables. Correlation analysis is used to "catch" whether there is a relation between two variables (properties, characteristics). What does correlation mean?  Here it is necessary to understand the relation as a kind of similarity in the "behavior of two characteristics". When one feature, characteristic increases does the other also increase? Or does it decrease? Or does it not change at all?

When calculating the correlation relation between two variables, we get a value whose absolute value (in some correlation coefficients we get both negative and positive values) is called the strength of the correlation, the strength of the relation between two variables. Very often there is a doubt how to interpret the size of such a coefficient, does it mean a strong or weak relationship?

Classification according to J.Guilford:

$|\rho|$=0 - no correlation

$0.0<|\rho|≤0.1$ - weak correlation

$0.1<|\rho|≤0.3$ - weak correlation 

$0.3<|\rho|≤0.5$ - average correlation

$0.5<|\rho|≤0.7$ - high correlation

$0.7<|\rho|≤0.9$ - very high correlation

$0.9<|\rho|<1.0$ - almost full correlation

$|\rho|=1$ - full correlation

Peterson's definition is the most commonly used:

According to this definition, the correlation coefficient $\rho(x,y)$ between two verialbles $x$ and $y$ is calculated as follows:

$$\rho(x,y)=\frac{cov(x,y)}{\sigma_x*\sigma_y}$$,

where:
- $cov(x,y)$ is covariance between $x$ and $y$
$$cov(x,y)=E(x∗y)−(E(x)∗E(y))$$
- E(x) expected value of x
- $\sigma_x$ standard deviation for varialble $x$

## How to plot data?:

```python
import matplotlib.pyplot as plt

# Given data
x = [1, 3, 6, 4, 15, 6, 10, 8, 9, 7, 11, 5, 13, 6, 5]
y = [33, 4, 21, 8, 28, 12, 16, 4, 17, 9, 11, 2, 55, 14, 7]

# Plotting the data
plt.figure(figsize=(10, 6))
plt.scatter(x, y, color='blue')
plt.title('Scatter Plot of x vs y')
plt.xlabel('x values')
plt.ylabel('y values')
plt.grid(True)
plt.show()
```

```python
import numpy as np

# Calculate the mean (expected value) and standard deviation for x and y
mean_x = np.mean(x)
mean_y = np.mean(y)
std_x = np.std(x, ddof=1)  # Sample standard deviation (N-1 in the denominator)
std_y = np.std(y, ddof=1)

# Calculate covariance
cov_xy = np.mean(np.multiply(x, y)) - (mean_x * mean_y)

# Calculate the correlation coefficient
correlation_coefficient = cov_xy / (std_x * std_y)

correlation_coefficient

```

# Exercise 2 - Correlation calculations

1. Determine the value of the correlation coefficient for your own data x and y

```python
# Code 1: Generate y values with a positive correlation to x using a predefined slope and intercept
np.random.seed(0)  # For reproducibility

# Generate a series of x values
x_positive = np.random.uniform(0, 50, 100)

# Define a slope and intercept for our linear relationship
slope = 2
intercept = -5

# Generate y values with some added noise for variance
noise = np.random.normal(0, 10, x_positive.shape)
y_positive = slope * x_positive + intercept + noise

# Calculate the positive correlation coefficient
positive_correlation_coefficient_1 = np.corrcoef(x_positive, y_positive)[0, 1]

# Plotting the positively correlated data
plt.figure(figsize=(10, 6))
plt.scatter(x_positive, y_positive, color='green')
plt.title('Scatter Plot of Positively Correlated Data')
plt.xlabel('x values')
plt.ylabel('y values')
plt.grid(True)
plt.show()

positive_correlation_coefficient_1
```

2. Find data for which the correlation coefficient will be negative (positive if you got a negative coefficient in the first step)

```python
# Code 2: Generate two normally distributed sets of data with a specified positive correlation

# Define the mean and covariance matrix with a specified correlation
mean = [0, 0]
cov = [[1, 0.8], [0.8, 1]]  # diagonal covariance, the off-diagonal values are the correlation

# Generate the normally distributed data
data = np.random.multivariate_normal(mean, cov, 100)
x_positive_2, y_positive_2 = data.T

# Calculate the positive correlation coefficient
positive_correlation_coefficient_2 = np.corrcoef(x_positive_2, y_positive_2)[0, 1]

# Plotting the positively correlated data
plt.figure(figsize=(10, 6))
plt.scatter(x_positive_2, y_positive_2, color='blue')
plt.title('Scatter Plot of Positively Correlated Data')
plt.xlabel('x values')
plt.ylabel('y values')
plt.grid(True)
plt.show()

positive_correlation_coefficient_2
```

1. Risks:
- Confusing Correlation with Causation: One of the most common risks is assuming that correlation implies causation. Just because two variables are correlated does not mean one causes the other.

- Presence of Outliers: Outliers can significantly affect the correlation coefficient, giving a misleading impression of the strength and direction of a relationship between variables.

- Linearity Assumption: The Pearson correlation coefficient measures linear relationships. If the relationship is non-linear, the correlation coefficient may underestimate the strength of the relationship.

- Restricted Range: If the data have a restricted range (e.g., measurements are taken under limited conditions), the correlation coefficient may not reflect the true relationship under a broader range of conditions.

2. Mitigation Strategies
- Causal Analysis: Use additional statistical methods to test for causality where appropriate, such as controlled experiments, longitudinal studies, or causal inference techniques.

- Outlier Analysis: Perform outlier detection and consider their treatment (e.g., removal or separate analysis) to ensure they do not unduly influence results.

- Range of Data: Ensure the data covers a sufficiently broad range to capture the variability in the variables of interest.
