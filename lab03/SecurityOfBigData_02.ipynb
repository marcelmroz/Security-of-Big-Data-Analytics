# Exercise 1 - Histogram


A histogram provides an approximate illustration of how numerical data is distributed. The term was originally introduced by Karl Pearson. To construct a histogram, the initial step involves "binning" or grouping the range of values, essentially dividing the entire value range into a series of intervals, and then tallying the number of values that fall within each interval. These intervals, known as bins, are typically defined as successive, non-overlapping ranges of a variable. The bins must be adjacent and are frequently (although not necessarily) of equal size. An example of a histogram is shown in Figure 1. The x-axis shows the ranges of values, while the y-axis shows the number of occurrences of numbers in a given range.

```python
import numpy as np
import pandas as pd
import os
import fnmatch
import matplotlib.pyplot as plt

# Set the display option for pandas
pd.options.display.max_rows = 9999

directory_path = './'

txt_files = fnmatch.filter(os.listdir(directory_path), '*.txt')

files_to_process = [
    "data_d01_01.txt",
    "data_d01_02.txt",
    "data_d01_03.txt",
    "data_d02_01.txt",
    "data_d02_02.txt",
    "data_d02_03.txt"
]

fig, axes = plt.subplots(nrows=2, ncols=3, figsize=(14, 8))  
axes = axes.ravel()

for idx, file in enumerate(files_to_process):
    if file in txt_files:
        data_file = pd.read_csv(file, header=None)  

        axes[idx].hist(data_file[0], bins=20, edgecolor='black') 
        axes[idx].set_title(f'Histogram of {file}')
        axes[idx].set_xlabel('Value')
        axes[idx].set_ylabel('Frequency')
    else:
        axes[idx].text(0.5, 0.5, f"{file} not found in directory.", 
                       horizontalalignment='center', 
                       verticalalignment='center',
                       transform=axes[idx].transAxes)
        print(f"{file} not found in directory.")
        axes[idx].set_title(f'{file} not found')
        axes[idx].set_xlabel('Value')
        axes[idx].set_ylabel('Frequency')

# Adjust the layout so the plots do not overlap
plt.tight_layout()
plt.show()
```

# Exercise 3 - Expected value, mode, median and standard deviation
For data obtained from different measurements, the mean, mode and median can be determined. To calculate that parameters we should use statistics library.

The expected value is the arithmetic mean of a large number of independently selected outcomes of a random variable. 

The median is the middle value of the data that we have ordered from smallest to largest. 

The mode is the value that occurs most frequently in the data set.

Standard deviation - the classic measure of variability, next to the arithmetic mean, the most widely used statistical concept. Intuitively speaking, the standard deviation tells you how widely the values of some quantity (for example, age, inflation, exchange rate) are scattered around its mean.

3. Calculate the parameters for the dataset read from files indicated by your instructor.

```python
import os
import fnmatch
import pandas as pd
import statistics

# Define the directory path where your files are located
directory_path = './'  # Update with the path to your files

# List of filenames provided by your instructor
files_to_process = [
    "data_d01_01.txt",
    "data_d01_02.txt",
    "data_d01_03.txt",
    "data_d02_01.txt",
    "data_d02_02.txt",
    "data_d02_03.txt"
]

# Process each file
for file_name in files_to_process:
    try:
        # Assuming the data is in a single column format, no header
        data = pd.read_csv(os.path.join(directory_path, file_name), header=None).squeeze("columns")
        
        # Calculate statistics
        x_mean = statistics.mean(data)
        x_median = statistics.median(data)
        try:
            x_mode = statistics.mode(data)
        except statistics.StatisticsError:
            x_mode = "No unique mode found"
        x_st_dev = statistics.stdev(data)
        
        # Print the results
        print(f"File: {file_name}")
        print(f"Mean: {x_mean}")
        print(f"Median: {x_median}")
        print(f"Mode: {x_mode}")
        print(f"Standard Deviation: {x_st_dev}")
        print("-----------")
    except FileNotFoundError:
        print(f"File {file_name} not found.")
    except Exception as e:
        print(f"An error occurred while processing {file_name}: {e}")
```

# Exercise  4 - Generation of data with a given distribution


The numpy library can be used to generate numbers with a given distribution (import numpy as np).
- uniform distribution: x = np.random.uniform(low, high, size) - (low - Lower boundary of the output interval, high - Upper boundary of the output interval, size - number of samples)
- normal distribution: x = np.random.normal(loc, sd, size) - loc - mean (centre) of the distribution, sd - Standard deviation (spread or width) of the distribution, size - number of samples.  
- exponential distribution: x =  np.random.exponential(scale, size) - scale - distribution parameter,  size - number of samples.

1 Generate sample sets of numbers with a given distribution and parameters (uniform, normal and expotential). Prepare histograms for different intervals. 

In the cell below is an example of how to generate data according to an exponential distribution with parameters 0.1 and 1000

```python
import numpy as np
x =  np.random.exponential(0.1, 10000)
#plt.hist(x,100)
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Set the number of samples
size = 10000

# Uniform distribution parameters
low = 0
high = 10

# Normal distribution parameters
loc = 0  # Mean
sd = 1   # Standard deviation

# Exponential distribution parameters
scale = 1  # Scale

# Generating the data
uniform_data = np.random.uniform(low, high, size)
normal_data = np.random.normal(loc, sd, size)
exponential_data = np.random.exponential(scale, size)

# Create a figure with subplots
fig, axs = plt.subplots(3, 1, figsize=(10, 15))

# Plot histograms
# Uniform Distribution
axs[0].hist(uniform_data, bins=50, color='blue', alpha=0.7)
axs[0].set_title('Uniform Distribution')
axs[0].set_xlabel('Value')
axs[0].set_ylabel('Frequency')

# Normal Distribution
axs[1].hist(normal_data, bins=50, color='red', alpha=0.7)
axs[1].set_title('Normal Distribution')
axs[1].set_xlabel('Value')
axs[1].set_ylabel('Frequency')

# Exponential Distribution
axs[2].hist(exponential_data, bins=50, color='green', alpha=0.7)
axs[2].set_title('Exponential Distribution')
axs[2].set_xlabel('Value')
axs[2].set_ylabel('Frequency')

# Display the plot
plt.tight_layout()
plt.show()
```
